Tratar los valores faltantes

Identificar los valores faltantes del dataset. Esto creará un DataFrame booleano en el que se indicará con True las celdas que tienen valores faltantes y False las que tienen datos.

```python
# Identificar valores faltantes en el dataset
valores_faltantes = pp.isna()
print(valores_faltantes)
```

Contar valores faltantes:

Para tener una idea de cuántos valores faltantes hay en cada columna, puedes usar sum() después de isna():

```python
faltantes_por_columna = pp.isna().sum()
print(faltantes_por_columna)
```

```python
pp = pp.dropna(subset=['Invoice amount gross home currency']) # Eliminar filas con valores faltantes en 'Columna_A'
print(pp)
```

```python
# Eliminar múltiples columnas a la vez
columnas_a_eliminar = ['Invoice referenced', 'Delivery', 'Tax %', 'Package', 'Contract customer ref', 'Requires PO', 'Agent parent', 'Advertiser parent', 'Advertiser holding', 'Advanced invoice number', '1', 'Rep name 3', 'Rep sales team 4', 'Rep sales team 5', 'Rep code 5', 'Rep name 5', 'Rep sales team 6', 'Rep code 6', 'Rep name 6', 'Rep sales team 7', 'Rep code 7', 'Rep name 7', 'Rep sales team 8', 'Rep code 8', 'Rep name 8', 'Rep sales team 9', 'Rep code 9', 'Rep name 9', 'Taxable', 'Consumption', 'Commission 4 Type', 'Commission 4 Gross/Net', 'Discount 1 Type','Discount 1','Discount 1 Gross/Net', 'Discount 1 amount', 'Discount 2 Type','Discount 2', 'Discount 2 Gross/Net','Discount 2 amount','Discount 3 Type','Discount 3', 'Discount 3 Gross/Net', 'Discount 3 amount', 'Discount 4 Type', 'Discount 4', 'Discount 4 Gross/Net', 'Discount 4 amount', 'Discount 1 File Code', 'Discount 2 File Code', 'Discount 3 File Code', 'Discount 4 File Code']
pp = pp.drop(columnas_a_eliminar, axis=1)
```

```python
print(pp)
```

Identificar valores atípicos

```python
# Lista de valores que deseas eliminar
valores_a_eliminar = ['Consumption', 'Contract instalment', 'Contract prebill', 'Independent deal instalment', 'Master deal instalment', 'Subdeal instalment']

# Crear una condición booleana que marque las filas a eliminar
condicion_eliminar = pp['Row type'].isin(valores_a_eliminar)

# Eliminar las filas que cumplan con la condición
pp = pp[~condicion_eliminar]
```
